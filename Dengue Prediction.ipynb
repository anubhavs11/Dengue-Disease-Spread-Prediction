```python
def computeCost(X,y,theta):
    n=len(X[0]);
    m=len(X);
    
    h = [0]*m
    
    for i in range(m):
        for j in range(n):
            h[i] += X[i][j]*theta[j]
    sum=0
    for i in range(m):
        sum+=h[i]-y[i]
        
    J = 1/(2*m) *sum
    return J
```

```python
X = [[1,2],[3,4]]
theta = [2,4]
y = [1,1]
J = computeCost(X,y,theta)
```
